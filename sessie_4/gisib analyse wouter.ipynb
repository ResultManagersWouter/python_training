```python
gdf_vrh["RELATIEVE_HOOGTELIGGING"] = gdf_vrh["RELATIEVE_HOOGTELIGGING"].fillna(0)
intersectie_verhardingen = gdf_vrh.overlay(gdf_vrh,how="intersection",keep_geom_type = True)
```

```python
# voorwaardes
mask_guid = intersectie_verhardingen.GUID_1 != intersectie_verhardingen.GUID_2
mask_area = intersectie_verhardingen.area > 5
mask_rel = intersectie_verhardingen.RELATIEVE_HOOGTELIGGING_1 == intersectie_verhardingen.RELATIEVE_HOOGTELIGGING_2
# totaal 
mask_total = (mask_guid) & (mask_area) & (mask_rel)
```

```python
# total bool indexing
mask_total
```

```python

intersection_total_mask = intersectie_verhardingen[mask_total]
```

```python
intersection_total_mask = intersection_total_mask.assign(
    guid_list= intersection_total_mask.apply(lambda v: sorted([v.GUID_1, v.GUID_2]), axis=1)
)
intersection_total_mask.drop_duplicates(subset=["guid_list"],keep="first").shape
```

```python
for idx,row in intersection_total_mask.iterrows():
    print(row.GUID_1,row.GUID_2)
```
